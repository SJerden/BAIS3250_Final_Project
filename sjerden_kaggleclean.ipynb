```python
import pandas as pd
```

```python
used_cars = pd.read_csv("used_cars.csv")
used_cars
```

```python
used_cars.isna().sum()
```

### Modifying data types so that data can be used for calculations later

```python
used_cars['new_price'] = used_cars['price'].str.replace("$","").str.replace(",","")
used_cars['new_price'] = used_cars['new_price'].astype("int")
```

```python
used_cars['new_mileage'] = used_cars['milage'].str.replace("mi.","").str.replace(",","")
used_cars['new_mileage'] = used_cars['new_mileage'].astype("int")
```

```python
used_cars['new_accident'] = used_cars['accident'].str.replace("None reported", "False").str.replace("At least 1 accident or damage reported", "True")
used_cars['new_accident'].astype("bool")
```

```python
used_cars
```

### Dropping redundant rows and renaming columns to match their predecessors

```python
used_cars = used_cars.drop(['milage','price', 'accident'], axis=1)
```

```python
used_cars = used_cars.rename(columns={
    'new_price': 'price',
    'new_mileage': 'mileage',
    'new_accident': 'accident'
})
```

```python
used_cars.to_csv("kaggle_cars_cleaned.csv")
```
