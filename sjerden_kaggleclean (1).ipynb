```python
import pandas as pd
```

```python
used_cars = pd.read_csv("used_cars.csv")
used_cars
```

```python
used_cars.isna().sum()
```

```python
used_cars.drop(['engine'], axis=1, inplace=True)
```

### Modifying data types so that data can be used for calculations later

```python
used_cars['price'] = used_cars['price'].str.replace("$","").str.replace(",","")
used_cars['price'] = used_cars['price'].astype("int")
```

```python
used_cars['milage'] = used_cars['milage'].str.replace("mi.","").str.replace(",","")
used_cars['milage'] = used_cars['milage'].astype("int")
```

```python
used_cars['accident'] = used_cars['accident'].str.replace("None reported", "False").str.replace("At least 1 accident or damage reported", "True")
used_cars['accident'].astype("bool")
used_cars.fillna("No", inplace=True)
```

```python
used_cars['data_year'] = 2023
```

```python
used_cars
```

```python
used_cars.to_csv("kaggle_cars_cleaned.csv")
```
